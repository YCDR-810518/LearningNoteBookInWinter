## .head()&.tail()查看部分行（默认前五行）
### 可以输入一个数字控制显示的行数

```python
import pandas as pd

# Popular mobile apps
app_data = {
  'app_name': [
    'YouTube', 'TikTok', 'Instagram', 'Spotify', 'Duolingo', 
    'Twitter', 'Headspace', 'Discord', 'Depop'
  ],
  'category': [
    'Video', 'Social Media', 'Social Media', 'Music', 'Education',
    'Social Media', 'Health', 'Communication', 'Shopping'
  ],
  'rating': [
    4.7, 4.6, 4.5, 4.6, 4.7,
    4.3, None, 4.7, 4.4
  ],
  'downloads_millions': [
    5000, 3000, 3500, 2000, None,
    1500, 500, 600, 200
  ]
}

apps = pd.DataFrame(app_data)

apps.head()
```

## .info()处理数据类型和缺失值

### 用法&用途

.info() 方法将显示有关 DataFrame 中每一列的信息。

```python
apps.info()
```

#### 9entries 表示数据集中有 9 行数据

#### non-null是非空值的意思，若与行数不一致，则说明有缺失值

#### Dtype是每一列数据的数据类型
**用于了解Frame中的数据类型&缺失的数据量**
* 十进制浮点数->float64
* 整数储存->int64
* 复杂对象/自定义对象->object


## .describe()统计摘要

.describe() 方法来获取每个数值列的汇总统计信息（均值、最小值、最大值、标准差等）：

```python
apps.describe()
```

### 添加include='all'可以查看非数值列的统计信息

```python
apps.describe(include='all')
```

## 查看特定的列

### 列(genre)

**类型：Series**
* 当我们选择单个列时，我们实际上得到的是一个 Pandas Series（而不是一个完整的 DataFrame）。
* 如果把 DataFrame 比作一张表格，那么 Series 就如同这张表格中的一列。它依然功能强大，只是范围更窄一些。

```python
print(type(apps['app_name']))
apps['app_name']
```

### 多个列的访问

```python
appNmae_cateGory_df = apps[['app_name','category']]
print(f'多个列的类型：{type(appNmae_cateGory_df)}')
print(f'单个列的类型：{type(apps['app_name'])}')
print(appNmae_cateGory_df)
```

## 删除df中的特定列.drop()方法

```python
## 尝试去除特定的列
apps_removed_rating_1 = apps.drop('rating', axis = 1)

print('去除特定的列后的详细列表')
print(apps_removed_rating_1)

print('去除特定的列后的详细信息')
apps_removed_rating_1.info()

print()
## 尝试去除特定的行
apps_removed_rating_0 = apps.drop(4, axis = 0)
print('去除特定的行后的详细列表')
print(apps_removed_rating_0)
print('去除特定的行后的详细信息')
apps_removed_rating_0.info()
```
