# pandas数据过滤

## filtering rows筛选行

````python
import pandas as pd

# Popular mobile apps
app_data = {
  'app_name': [
    'YouTube', 'TikTok', 'Instagram', 'Spotify', 'Duolingo', 
    'Twitter', 'Headspace', 'Discord', 'Depop'
  ],
  'category': [
    'Video', 'Social Media', 'Social Media', 'Music', 'Education',
    'Social Media', 'Health', 'Communication', 'Shopping'
  ],
  'rating': [
    4.7, 4.6, 4.5, 4.6, 4.7,
    4.3, None, 4.7, 4.4
  ],
  'downloads_millions': [
    5000, 3000, 3500, 2000, None,
    1500, 500, 600, 200
  ]
}

apps = pd.DataFrame(app_data)

print(apps)
````

### 行过滤的语法
* df[df['用作判断的列'] 判断语句]

````python
apps_filted_rating = apps[apps['rating'] >=4.6]

# 查看内层
apps_rating =apps['rating'] >=4.6
print('内层的相关信息')
apps_rating.info()
print('内层的相关描述和全部数据内容')
apps_rating.describe()
print(apps_rating)
# 过滤后的内容
print('过滤后的内容')
print(apps_filted_rating)
````

#### 由上面的程序可知，要筛选 DataFrame，我们使用方括号内的布尔 Series (apps_rating)。这样就可以只保留条件为 True 的行。

----
### 多次筛选（使用多个判断语句）

#### 注意，在判断的项与项之间要用```&```和```|```号连接！

````python
popular_apps = apps[
    ((apps['rating']>=4.6) & (apps['rating'] != None))&
    ((apps['downloads_millions']>=2000) & (apps['downloads_millions'] != None))
    ]
print(popular_apps)
````
