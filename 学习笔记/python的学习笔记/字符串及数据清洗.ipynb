## 常用的字符串函数

### capitalize()

1. 首字符会转换成大写，其余字符会转换成小写。

2. 首字符如果是非字母，首字母不会转换成大写，会转换成小写。

```python
str = 'hello python'
str.capitalize()       
```

### center(居中的宽度,填充的字符(默认为空格))

```python
# 1、如果 width 小于字符串宽度直接返回字符串，不会截断:
str = 'GDUT'
print('str.center(3)')
print(str.center(3))

# 2、fillchar 默认是空格
print('str.center(10)')
print(str.center(10))

# 3、第二个参数决定了填充的字符
print('str.center(10,"*")')
print(str.center(10,'*'))

# 4、fillchar 只能是单个字符
print ("str.center(40, '?!') : \n", str.center(40, '?!'))
```

### find()方法

可以用来查找要删除的字符
* **语法：**：str.find(str, beg=0, end=len(string))

```python
str1 = "Runoob example....wow!!!"
str2 = "exam";
 
print (str1.find(str2))
print (str1.find(str2, 5))
print (str1.find(str2, 10))
```
