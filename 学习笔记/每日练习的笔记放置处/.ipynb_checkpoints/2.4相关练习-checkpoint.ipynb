* 截获一条杂乱情报：```" Agent:007_Bond; Coords:(40,74); Items:gun,money,gun; 
Mission:2025-RESCUE-X "```
* 请运用所有所学知识清洗数据：利用 String 方法去除干扰空格；利用 Set 帮特工去除重复装
备；利用 Slicing 截取核心任务代号；利用 Tuple锁定坐标；最后将所有信息归档进一个 Dict 
档案中

## 1.清洗数据

````python
message = ' Agent:007_Bond; Coords:(40,74); Items:gun,money,gun; Mission:2025-RESCUE-X '

# 除去空格
message = message.strip(' ')
print(message)

# 按照空格进行第一次切分
messageList = message.split(' ')
print(messageList)

# 再除去多余分号
i = 0
for i in range(len(messageList)):
    messageList[i] = messageList[i].strip(';')
    i += 1
print(messageList)

# 进行二次切分，创建字典
messageDict = {}
for slip in messageList:
    templist1 = slip.split(':')
    print(templist1)
    # 二次分割
    # 下面这句太c风格了
    #if templist1[1].find('(') >= 0 and templist1[1].find(')')>=0:
    if '(' in templist1[1] and ')' in templist1[1]:
        # 先替换
        cleanedList = templist1[1].strip('(')
        cleanedList = cleanedList.strip(')')
        cleanedList = cleanedList.replace('()','|')
        templist2 = cleanedList.split('|')
        print(templist2)
    else:
        templist2 = templist1[1].split(',')
        print(templist2)
    
    # 将值放入集合中进行再清洗
    templist2 = list(set(templist2))
    print(f'最终清洗完成的键值：\n{templist2}')
    messageDict[templist1[0]] = templist2
print(messageDict)

````

## 嵌套类型数据的清洗方式

1. 简单版：将不对称的分隔符'(',')'化为对称分隔符
```Python
data = "(a,b),(c,d),(e,f)"

# 1. 先去掉最外层的括号，并把中间的 "),(" 替换成统一的分隔符（比如 |）
cleaned = data.strip('()').replace('),(', '|')

# 2. 按照 | 切分出每一组
groups = cleaned.split('|')

# 3. 再对每一组进行内部切分
result = [tuple(item.split(',')) for item in groups]

print(result)
# 输出: [('a', 'b'), ('c', 'd'), ('e', 'f')]
```
2. 正则表达式法（高效！）
```python
# 假设 value 现在是 "(a,b),(c,d),(e,f)"
if value.startswith('(') and value.endswith(')'):
    # 使用正则表达式快速提取
    import re
    pairs = re.findall(r'\(([^)]+)\)', value)
    # 存成嵌套列表或元组
    clean_value = [tuple(p.split(',')) for p in pairs]
else:
    clean_value = value.split(',')
```
